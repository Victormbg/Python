```python
## APS 1 - Estatística Computacional - BS - CET0018N - 2022-1
## Autores:
## Marcos Paulo C. Tavares - 19200427
## Victor Manuel de Barros Garcia - 20102283

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def CalcularTotalArray(array):
    total = 0
    for i in array:
        total += i
    return total

def FrequenciaAcumulada(fi):
    for index, i in enumerate(fi):
        if index == 0:
            array_FrequenciaAcumulada = np.array([i])
        else:
            arrayCortado_FrequenciaAbsoluta = fi[0:index]
            result = CalcularTotalArray(arrayCortado_FrequenciaAbsoluta) + i
            array_FrequenciaAcumulada = np.append(array_FrequenciaAcumulada, result)
    return array_FrequenciaAcumulada

def FrequenciaRelativaAbsoluta(fi):
    for index, i in enumerate(fi):
        if index == 0:
            total_FrequenciaAbsoluta = CalcularTotalArray(fi)
            result = ((i/total_FrequenciaAbsoluta)*100)/100
            array_FrequenciaRelativaAbsoluta = np.array([result])
        else:
            total_FrequenciaAbsoluta = CalcularTotalArray(fi)
            result = ((i/total_FrequenciaAbsoluta)*100)/100
            array_FrequenciaRelativaAbsoluta = np.append(array_FrequenciaRelativaAbsoluta, result)
    return array_FrequenciaRelativaAbsoluta

def FrequenciaRelativaAcumulada(fri):
    for index, i in enumerate(fri):
        if index == 0:
            array_FrequenciaRelativaAcumulada = np.array([i])
        else:
            arrayCortado_FrequenciaRelativaAcumulada = fri[0:index]
            result = CalcularTotalArray(arrayCortado_FrequenciaRelativaAcumulada) + i
            array_FrequenciaRelativaAcumulada = np.append(array_FrequenciaRelativaAcumulada, result)
    return array_FrequenciaRelativaAcumulada
```

```python
# Notas
notas = np.array([3.0,4.0,5.0,6.0,7.0,8.0,9.0])
# Array de Frequencia Abosulta com as repetições de notas por alunos
fi = np.array([1, 3, 4, 6, 5, 4, 2])

#Fi = [1,4,8,14,19,23,25]
Fi = FrequenciaAcumulada(fi)
#fri = [0.04,0.12,0.16,0.24,0.20,0.16,0.08]
fri = FrequenciaRelativaAbsoluta(fi)
#Fri = [0.04,0.16,0.32,0.56,0.76,0.92]
Fri = FrequenciaRelativaAcumulada(fri)

print("CHECANDO OS DADOS:",
      "\nNotas: ",notas,
      "\nFrequencia Absoluta - fi: ",fi,
      "\nFrequencia Acumulada - Fi: ",Fi,
      "\nFrequencia Relativa Absoluta - fri: ",fri,
      "\nFrequencia Relativa Acumulada - Fri: ",Fri,
      )
```

```python
# Tabela
tabela = {'Notas': notas,
          'N° alunos(fi)': fi,
          'Fi': Fi,
          'fri': fri,
          'Fri': Fri
        }

tabela_df = pd.DataFrame(tabela)
```

```python
#Frequência Absoluta ( fi )
grafico_fi_df = pd.DataFrame({'notas':notas, 'fi':fi})
ax = grafico_fi_df.plot.bar(x='notas', y='fi', rot=0)
plt.title('Frequência - fi', fontsize=16)
plt.grid(axis = 'y')
plt.show()
```

```python
#Frequência Acumulada ( Fi )
grafico_Fi_df = pd.DataFrame({'notas':notas,'Fi': Fi})
lines = grafico_Fi_df.plot.line(x='notas', y='Fi', rot=0, marker = 'o')
plt.title('Frequência Acumulada - Fi', fontsize=16)
plt.grid(axis = 'y')
plt.show()
```

```python
#Frequência relativas absolutas ( fri )
grafico_fri_df = pd.DataFrame({'notas':notas,'fri': fri})
lines = grafico_fri_df.plot.line(x='notas', y='fri', rot=0, marker = 'o')
plt.title('Frequência Relativas Absolutas - fri', fontsize=16)
plt.yticks(fri)
plt.grid(axis = 'y')
plt.show()
```

```python
#Frequência relativas absolutas ( Fri )
grafico_Fri_df = pd.DataFrame({'notas':notas,'Fri': Fri})
lines = grafico_Fri_df.plot.line(x='notas', y='Fri', rot=0, marker = 'o')
plt.title('Frequência Relativas Acumulada - Fri', fontsize=16)
plt.yticks(Fri)
plt.grid(axis = 'y')
plt.show()
```

```python
#Todas as Frequências
graficoTotal_df = pd.DataFrame({'fi':fi,'Fi': Fi,'fri': fri,'Fri': Fri}, index=notas)
lines = graficoTotal_df.plot.line(rot=0, marker = 'o')
plt.title('Grafico das Frequências', fontsize=16)
plt.xlabel('Notas', fontsize=10)
plt.grid(axis = 'y')
plt.show()
```
